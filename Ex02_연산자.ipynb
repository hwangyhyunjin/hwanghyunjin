```python
num1 = int(input("python 점수 입력>> "))
num2 = int(input("머신러닝 점수입력>> "))
num3 = int(input("딥러닝 점수 임력>> "))
total = num1+num2+num3
avg = total/3
print("합계 : {}".format(num1+num2+num3))
print("평균 : {}".format((num1+num2+num3)/3))
print("합계 : {}".format(total))
print("평균 : {}".format(avg))
```

비교 연산자

```python
#뭔가를 비교하게, 조건이 되어 질 수 있음
# >, ==, <, <=, >=,!=
a = 3
b = 7
print(a>b) #False ->파이선에서는 꼭 대문자 F로 사용해야함, 소문자로 사용시 변수명으로되버림
print(a<b)
print(a<=b)
print(a>=b)
print(a==b)
print(a!=b)
```

```python
import random
a=random.randint(1,10)
b=random.randint(1,10)
print(f"a:{a}, b:{b}")

print(a<b)
print(a>b)
print(a==b)
print(a!=b)
```

```python
#not 논리 값을 뒤집는 역활
a=3
b=7
a<b #True
not a<b
```

```python
a == b #False
not a == b
```

```python
# and, or 조건들을 연결시킴
a<b and a==b #둘다 True ->True, 한개라도 False ->False
```

```python
a<b or a==b #한개라도 True ->True
```
