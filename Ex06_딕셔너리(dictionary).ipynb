### 딕셔너리 값 수정

```python
dic_test={"노래제목":"아무노래"}
dic_test
```

```python
dic_test["노래제목"]="신호등"
dic_test
```

```python
#리스트 형태!!
play_list=[{"노래제목":"사랑은 늘 도망가","가수":"임영웅"},
{"노래제목":"신호등","가수":"이무진"},{"노래제목":"문득","가수명":"비오"}]
```

```python
#play_list for 문 돌리기
#각각 딕셔너리로 구분 됨
for s in play_list:
    print(s)
```

```python
#노래 제목만 출력
# play_list -> 현재 [] 로 되어 있는 리스트 형태임
# 데이터 접근 위해서 딕셔너리로 구분해줘야함
for s in play_list:
    print(s["노래제목"])
```

딕셔너리 데이터 접근

- dict.["키값"] : 없으면 에러남 
- dict.get("키값") : 없으면 에러x , None(null)반환

```python
for s in play_list:
    print(s.get("가수"))
    print(s.get("가수명")) # 없으면 에러x , None(null)반환 
```
