### 함수정의

```python
#더하기 기능 (함수)
#매개변수2개: 숫자 2개
#더하는 수행코드 
#더해진 결과 반환
def number_sum(num1,num2):
    result=num1+num2
    return result
```

#### 매개변수 
- 기본값 설정

```python
#지수 연산하는 함수
```

```python
#정의
def power_of_N(num,power=2): #power=2 2가 디폴트 값으로 설정됨
    return num**power
```

```python
#함수 호출
power_of_N(3) # 매개변수를 하나만 입력했지만 오류나지 않고 num=3, power=디폴트 값으로 실행
```

```python
#3의 3제곱
power_of_N(3,3) # 기본값과 별개로 매개변수입력해도 됨
```

#### return 정리

```python
#1. return 함수안에서 한번만 실행
#2. return 을 만나면 함수 끝남
#3. return result,result2 튜플로 묶어서 반환
```

```python
def add_sub(num1,num2):
    return num1+num2, num1-num2
```

```python
add_sub(10,7) #튜플로 반환
```

```python
add,sub = add_sub(10,7)
print(add,sub)
```
