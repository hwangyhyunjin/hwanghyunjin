## while문

```python
#while 조건식:
    #실행문장
```

```python
#파이썬 최고 13번 출력해보기!
num=1
while num<=13:
    print(f"{num}, 파이썬 최고!")
    num+=1
```

```python
#while 반복을 멈추게하는 키워드 :break
```

```python
num =1
while True:
    print("실행~!!!")
    if num==2:
        break
    num+=1
```

```python
while True:
    num1=int(input("첫 번째 정수 입력>> "))
    num2=int(input("두 번째 정수 입력>> "))
    num3= num1+num2
    if num1==0 and num2==0:
        break
    print("두 정수의 합: ",num3)
print("프로그램이 종료되었습니다")
```

## for문

```python
#for문 예시
list_food =["서브웨이","돈까스","쭈꾸미","찜닭","떡볶이"]
len(list_food) # len : 길이 구하는 함수 
for food in list_food:
    print(food)
    print("안녕하세요!") #for문 리스트 길이와같이 5번 돌기때문에 !! 5번 출력
```

```python
#for 예시 -문자열
s="정림이네 국밥"
for c in s:
    print(c, end="") #print 속성안에 end=\n 포함되어 있어서 자동 줄바꿈됨
```

```python
# for 예시 - tuple
tuple_food=("햄버거","피자","치킨")
cnt= 0
for i in tuple_food:
    cnt+=1
    print(cnt,i)
```

```python
score_list=[90,45,70,60,55]
stu=0
for score in score_list:
    stu+=1
    if score >= 60:
         print(f"{stu}번 학생은 합격입니다.")
    else:
        print(f"{stu}번 학생은 불합격입니다.")
      
```

```python
# lange 함수 사용
score_list=[90,45,70,60,55]
for i in range(0,len(score_list),1):
    if score_list[i]>=60:
        print(f"{i+1}번 학생은 {score_list[i]}점 합격입니다.")
    else:
        print(f"{i+1}번 학생은 불합격입니다.")
```

```python
list2=[4,5,2,1,99,15,2,7,27]
max_data=list2[0]
min_data=list2[0]
for i in range(0,len(list2),1):
    if list2[i] > max_data:
        max_data=list2[i]
    if list2[i] < min_data:                                                                                                                                                                                                                                                                                                                                                                                                                                                                                
        
        min_data=list2[i]
print("가장 큰 수:",max_data)
print("가장 작은 수:",min_data)
```

```python
list2=[4,5,2,1,99,15,2,7,27]
max_data=list2[0]
for i in list2:
    if max_data<i:
        max_data=i
print(max_data)
```

```python
list_score=[["병관",80],["선재",90],["현지",85],["현준",100]]
#4명의 평균 점수를 구하시오 for 문은 사용하여!
total=0
for i in list_score:
    total+=i[1]
print(f"평균값은 {total/len(list_score)}입니다")    
```

```python
#list_score 85 이상인 사람의 이름 출력
print("85점 이상인 사람은",end=" ")
for i in list_score:
    if i[1]>=85:
        print(i[0], end=" ")
```

```python
#85점 이상인 사람이름을 저정해라
lst4=[]
print("85점 이상인 사람은",end=" ")
for i in list_score:
    if i[1]>=85:
        lst4.append(i[0])
print(lst4)
```

#### pickle 
파이썬 자료형 그대로 반영해서 데이터를 저장하는 도구(모듈)

```python
import pickle
```

```python
with open('movie_review.pkl','rb') as f:
    data=pickle.load(f)
```

```python
data
# [[리뷰문장,긍정1/부정0],[리뷰문장,긍정1/부정0].... ]
```

```python
# for문 이용해서 데이터 접근하는것
#리뷰만 출력해보자
for i in data:
    print(i[0])
```

```python
#전체 리뷰 갯수를 확인해보자 : 리스트트의 길이만 확인 (len)
len(data)
```

```python
#긍정,부정 출력해보자
for i in data:
    print(i[1])
```

리뷰 길이의 평균값



```python
total=0
for i in data:
  total += len(i[0]) #리뷰 문장 길이 누적
print(f"리뷰의 평균 길이는 '{total/len(data)}'입니다.") 
```

```python
p=0
n=0
for i in data:
    if i[1]==0:
        n+=1
    if i[1]==1:
        p+=1
print(f"긍정{p}개")
print(f"부정{n}개")
```

```python
len_max= len(data[0][0]) #임의로 최댓값 설정
cont=""  # 빈 문자열로 선언
for i in data:
    if len(i[0])>len_max:
        len_max=len(i[0])
        cont=i[0]
print(f"리뷰 최대길이는 '{len_max}'입니다\n{cont}")
```

```python
ㅋ_cnt=0
for i in data:
    #i[0] 번째 문자 안에 ㅋ이 있느냐 없으냐를 판단
    if "ㅋ" in i[0]:
        ㅋ_cnt+=1
print(f"'ㅋ'이 포한된 리뷰는 {ㅋ_cnt}개 입니다")
```

```python
input_cnt=0
input_str=input("검색할 데이터 입력>> " )
for i in data:
    if input_str in i[0]: # 맴버연산자->'찾고싶은문자' in 찾을대상
        input_cnt+=1
print(f"'{input_str}'이 포한된 리뷰는 {input_cnt}개 입니다")
```
