### pos tagging 활용

```python
data = '먹는다 먹다 먹었다 이쁘다 아름답다 사진 모자'
kkma.morphs(data)
```

```python
kkma.pos(data)
```

```python
kkma.tagset
```

```python
d = pd.DataFrame(kkma.pos(data), columns = ['morph', 'tag'])
d.set_index('tag',inplace=True)
d
```

```python
d.loc[['VV','VA','NNG']]['morph'].values
```

```python
def myTokenizer2(text):
    d = pd.DataFrame(kkma.pos(text), columns = ['morph', 'tag'])
    
    if ('VV' in d.tag.values) | ('VA' in d.tag.values) | ('NNG' in d.tag.values):
        return d[(d['tag'] == 'VV') | (d['tag'] == 'NNG') | (d['tag'] == 'VA')]['morph'].values
    else:
        return ' '
```

```python
tfi_vect = TfidfVectorizer(tokenizer = myTokenizer2)
tfi_vect.fit(X_train[:10000])
```

```python
tfi_vect.vocabulary_
```

```python
X_train_tf = tfi_vect.transform(X_train[:10000])
X_test_tf = tfi_vect.transform(X_test[:10000])
```

```python
#학습할 모델 import
from sklearn.linear_model import LogisticRegression
```

```python
logi = LogisticRegression()
logi.fit(X_train_tf, y_train[:10000])
logi.score(X_test_tf, y_test[:10000])
```

```python
# 데이터 프레임 형태로
voc = pd.DataFrame(tfi_vect.vocabulary_.keys(), 
                   index = tfi_vect.vocabulary_.values())
voc.sort_index()
```

```python
# 가중치 값 확인
logi.coef_
```

```python
#기존 데이터 프레임 형태라 시리즈 형태로 변경
voc.sort_index()[0]
```

```python
# 데이터 프레임 형태로
result = pd.DataFrame(logi.coef_.T, index = voc.sort_index()[0], columns=['w'])
result.sort_values(by = 'w')
# 부정 음수 , 긍정 양수
```
