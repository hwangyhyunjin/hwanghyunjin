### 문자열 데이터 접근
- 인덱싱 : 무엇인가를 하나하나 '가리킴' 
- 슬라이싱 : 무엇인가를 '잘라냄'(범위)

문자열 인덱싱

```python
jy_info = "My name is JY"
# M 데이터에 접근 -> M 이 가지고 있는 인덱스 번호 불러
print(jy_info[0])
# i 데이터에 접근 -> i 가 위치한 인데스 번호 불러
print(jy_info[8])
```

문자열 슬라이싱

```python
# 대상문자열[시작인덱스번호:끝인덱스번호+1]
```

```python
#My 접근해보기
print(jy_info[0:2])
```

```python
print(jy_info[0:1])
```

```python
print(jy_info[3:7])
print(jy_info[-2:]) #끝까지 슬라이싱은 끝값 생략가능
print(jy_info[:7]) #처음부터 슬라이싱은 시작값 생략가능
```

```python
# 슬라이싱 - 처음부터 끝까지 슬라이싱
# 문자열대상[시작값:끝값+1:증갑값]
jy_info[:]  # 인덱스 번호 둘다 생략가능
```

```python
jy_info[0:14:2]
# 문자열 뒤에서 부터 앞으로 출력하기
jy_info[::-1]
```

```python
day = "2020년 3월 3일의 날씨는 맑음입니다."
print("날짜: ",day[:11])
print("날씨: ",day[-6:-4])
```

문자열 관련 함수

```python
# 문자열에 포함된 문자 개수 세기
# 소문자 대문자 구별됨
str1="Hi, My name is HJ"
#str1.count("i")
str1.count("h")
```

```python
# 위의 대소문자 구별 없이 count 하고싶을 경우
# 문자열을 소문자 -> 대문자 
str1.upper()
# 대문자 -> 소문자
str2 = str1.lower()
str2.count("h")
```

```python
# find , index  문자 위치 알려주기
str1.find("i") #문장안에서 첫번째로 만나는 i의 인덱스 반환
str1.find("z") # 문장안에 없는 단어 찾기 -> -1 반환,에러나지 않음
```

```python
str1.index("i")
str1.index("z") # 에러 발생,
```

```python
# 'Hi, ' 라고 확인되는데 공백을 없애주는 함수 
# rstrip() 함수 사용시 오른쪽 공백제거 -> 'Hi,'  공백 사라짐
str1[:4].rstrip()
```

```python
# lstrip() 함수 사용시 왼쪽 공백제거
str1[3:].lstrip()
```

```python
s3 = "   빅데이터 화이팅!   "
s3 # '   빅데이터 화이팅!   ' 출력
s3.stri p() # '빅데이터 화이팅!' 출력 ,공백제거
```

```python
# replace("문자1", "문자2") -> 대체하다, 문자열1을 문자열2로 바꿔줌
s3.replace("화이팅","파이팅!!")
```

```python
# 문자열 나누기
s4 ="오늘도 고생하셧습니다, 조심히 들어가세요~"
s4.split(",") # 리스트로 반환 ['오늘도 고생하셧습니다', ' 조심히 들어가세요~']
s4.split(",")[1].strip() # '조심히 들어가세요~'  반환
```
