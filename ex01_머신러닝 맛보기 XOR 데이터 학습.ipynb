```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 데이터 셋 구축

```python
X = np.array([[0,0],         #문제에 해당하는 데이터는 X 변수에 넣는다.
              [0,1],         #대문자 X 
              [1,0],
              [1,1]])


y = np.array([0, 1, 1, 0])   #답에 해당하는 데이터는 y(소문자) 변수에 넣는다.
```

## 지도학습
- 답을 명시해서 학습하는 방법

### 모델생성

```python
from sklearn.neighbors import KNeighborsClassifier #분류
```

```python
#객체생성후 model이라는 변수에 넣어준다!
model = KNeighborsClassifier(n_neighbors=1) 
```

###  모델학습

```python
model.fit(X,y)  #공부시키기(학습)
```

###  모델예측

```python
X.shape
```

```python
#학습한 문제와 같은 형태로 넣어줘야함(중요!!!)
pre = model.predict([[0,0],
                    [1,1],
                    [0,0],
                    [1,0]])  
```

```python
pre
```

```python
#학습한 문제와 같은 형태로 넣어줘야함(중요!!!)
pre = model.predict([[0,0]])
```

```python
pre
```

###  모델평가

```python
model.score(X, y)
# 1.0 -> 100% 맞춘다는 의미
```
