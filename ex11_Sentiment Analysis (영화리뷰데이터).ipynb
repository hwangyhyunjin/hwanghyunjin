### 토큰화 & 특성추출(수치화) - BOW
- 텍스트를 수치화 한다

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
words = ['I love you', 'Hello my name is hyunjin', 'good choice', '현우 천재']
```

```python
vect = CountVectorizer()
```

```python
#한글을 숫자 데이터로 바꿔보쟈 (토큰화)
#가지고 있는 모든 데이터를 사용해서 단어사전 제작
vect.fit(words)
```

```python
#vocabulary_ : 단어사전 확인
print(vect.vocabulary_)
```

```python
vect.transform(words).toarray()
```

### 학습
