### 조건문

if

```python
if True:
    print("조건문 안 실행문장")
print("조건문 밖 실행문장")
```

```python
if False:
    print("조건문 안 실행문장")
print("조건문 밖 실행문장")
```

```python
a=1000
b=500
if a>b:
    print("a가 더 크다")
```

```python
a=1000
b=5000
if a>b:
    print("a가 더 크다")
else:
    print("b가 더 크다")
```

```python
# 조건이 더 많아졌을 경우
# if elif else
a=5000
b=5000
if a>b:
    print("a가 더 크다")
elif a<b:
    print("b가 더 크다")
else:
    print("a와 b는 같다")
print("조건문 종료")
```

```python
score =int(input("점수 입력 >> "))
if score>=90:
    print(score,"점은 A학점 입니다.")
elif score>=80:
    print(score,"점은 B학점 입니다.")
elif score>=70:
    print(score,"점은 C학점 입니다.")
elif score>=60:
    print(score,"점은 D학점 입니다.")
else:
    print(score,"점은 F학점 입니다.")
```

```python
score =int(input("점수 입력 >> "))
grade = ""
if score >=90:
    grade = "A"
elif score >=80:
    grade = "B"
elif score >=70:
    grade = "C"
elif score >=60:
    grade = "D"
else:
    grade = "F"
    
print(f"{score}점은 {grade}학점입니다.")
```

```python
countOfMask = int(input("마스크 개수 입력 >> "))

if countOfMask%8 == 0 :
   print("8개들이 포장지의 개수 :",countOfMask//8 )
if countOfMask%8/5 != 0:
    print("5개들이 포장지의 개수:", int(round((countOfMask%8)/5,0) )) 
if  countOfMask%8%5 != 0:
    print("5개들이 포장지의 개수:",int(round((countOfMask%8)/5+1,0) ) )
```

```python
countOfMask = int(input("마스크 개수 입력 >> "))

if countOfMask >=8:
    #//8들이로 나눔
    #8개들이로 나눈 나머지값 -> 5개들이 +1경우가 생김
    a=countOfMask//8
    b=(countOfMask%8)//5
    c=(countOfMask%8)%5
    if c>0:
        b += 1
    print("8개들이 포장지 개수:",a)
    print("5개들이 포장지 개수:",b)    
else:
    #//5들이로 나눔
    #5들이로 나눈 나머지값 ->5개들이 +1
    a=countOfMask//5
    b=countOfMask%5
    if b>0:
        a += 1
    print("5개들이 포장지 개수:",a)
```
