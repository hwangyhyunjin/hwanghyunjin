### 튜플(tuple) 

```python
# 파이썬 자료 구조 중 하나 
# 순서가 있는 대상
# 소괄호 ()로 작성, 내부요소는 , 콤마로 구분 
# 추가,삭제,수정 불가능
```

```python
#튜플 선언
a=() # 빈 튜플 선언
b=(1,2,3) # 정수형 데이터 담은 튜플
c=("a","b","c") #문자열 담은 튜플
d=(1,2,"a","b") #자료형 상관없이 선언 가능
e=(1,2,("a","b")) #이중(다중) 튜플 선언가능
```

```python
# 공통점
#타입(자료형)과 상관없이 인련의 요소(값 ,콤마로 관리)를 갖을수 있다 
#요소의 순서를 관리한다(인덱스번호)
```

```python
lst=["딸기","바나나","수박","체리","포도"]
str1=input("검색할 문자를 입력하세요>> ")
str1 in lst #결과가 true 또는 false로 출력 되기때문에 조건문에 바로 대입하면 됨
if str1 in lst: 
    print(f"{str1}는 리스트에 {lst.index(str1)}번째 인덱스에 들어 있습니다.")
else:
    print(f"{str1}는 리스트에 들어있지 않습니다.")
```
